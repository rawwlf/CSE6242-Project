```python
import folium
import openrouteservice
```

```python
ORS_API_KEY = "5b3ce3597851110001cf62483720ba554ecb472ca9f909525b366441"
client = openrouteservice.Client(key=ORS_API_KEY)
```

```python
# Define coordinates (longitude, latitude) - Example: Berlin to Munich
start = [-73.789955, 40.643670]   # JFK
end = [-74.016928,40.703517]    # Ellis Tickets
```

```python
route = client.directions(
    coordinates=[start, end],
    profile="driving-car",  # Other options: 'cycling', 'walking'
    format="geojson"
)
```

```python
print(route)  # Print the route information for debugging
```

```python
# Extract route geometry (line coordinates)
route_coords = route["features"][0]["geometry"]["coordinates"]
route_coords = [(lat, lon) for lon, lat in route_coords]  # Reverse for Folium
```

```python
# Create Folium map centered at the start location
m = folium.Map(location=[start[1], start[0]], zoom_start=10)
```

```python
# Add route line to the map
folium.PolyLine(route_coords, color="blue", weight=5, opacity=0.8).add_to(m)
```

```python
# Add markers for start and end points
folium.Marker([start[1], start[0]], popup="Start", icon=folium.Icon(color="green")).add_to(m)
folium.Marker([end[1], end[0]], popup="End", icon=folium.Icon(color="red")).add_to(m)
```

```python
# Save the map to an HTML file and display
m.save("route_map.html")
m
```
